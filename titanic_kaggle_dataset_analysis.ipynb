```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

#Config do gráfico!!!
sns.set(style="whitegrid")
```

```python
# Quantidade de sobreviventes e não sobreviventes de acordo com a classe de embarque
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=df_titanic, palette='viridis')
plt.title('Quantidade de Sobreviventes e Não Sobreviventes por Classe de Embarque')
plt.xlabel('Classe de Embarque')
plt.ylabel('Quantidade')
plt.legend(title='Sobreviveu', loc='upper right')
plt.show()
```

Gráfico 1: A distribuição de sobreviventes e não sobreviventes varia entre as classes de embarque. As classes superiores (1ª classe) tiveram uma taxa de sobrevivência maior em comparação com a 3ª classe, provavelmente por sua melhor estrutura e localização no Titanic.

```python
# Relação entre idade e valor pago na passagem de acordo com a coluna de sexo
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Sex', data=df_titanic, palette='deep')
plt.title('Relação entre Idade e Valor Pago na Passagem por Sexo')
plt.xlabel('Idade')
plt.ylabel('Valor da Passagem')
plt.legend(title='Sexo')
plt.show()
```

Gráfico 2: A relação entre idade e valor pago na passagem por sexo mostra que, em média, passageiros do sexo feminino tendem a pagar mais pelas passagens, o que pode estar relacionado à classe de embarque.

```python
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df_titanic, palette='coolwarm')
plt.title('Relação entre Idade e Valor Pago na Passagem por Sobrevivência')
plt.xlabel('Idade')
plt.ylabel('Valor da Passagem')

scatter.legend(title='Status')

plt.show()

# 0 sendo que morreu e 1 que sobreviveu
```

Gráfico 3: A relação entre idade e valor pago na passagem por sobrevivência indica que passageiros que pagaram mais pelas passagens (provavelmente em classes superiores) tinham maior probabilidade de sobreviver.

```python
# Relação entre idade e valor pago na passagem de acordo com a coluna de embarcados.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Embarked', data=df_titanic, palette='muted')
plt.title('Relação entre Idade e Valor Pago na Passagem por Porto de Embarque')
plt.xlabel('Idade')
plt.ylabel('Valor da Passagem')
plt.legend(title='Porto de Embarque')
plt.show()
```

Gráfico 4: A relação entre idade e valor pago na passagem por porto de embarque revela padrões distintos de preços e idades dependendo do porto de embarque, com Cherbourg (C) tendo muitos passageiros que pagaram altos valores, ou seja, Cherboug era uma região mais rica.

```python
# Relação entre idade e valor pago na passagem de acordo com a coluna pcclass
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Pclass', data=df_titanic, palette='Set1')
plt.title('Relação entre Idade e Valor Pago na Passagem por Classe de Embarque')
plt.xlabel('Idade')
plt.ylabel('Valor da Passagem')
plt.legend(title='Classe de Embarque')
plt.show()
```

Gráfico 5:
A relação entre idade e valor pago na passagem por classe de embarque mostra que a 1ª classe tinha passageiros que pagavam muito mais, independentemente da idade.

```python
# Quantidade de Masculinos e Femininos
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', data=df_titanic, palette='pastel')
plt.title('Quantidade de Passageiros por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Quantidade')
plt.xticks(ticks=['male', 'female'], labels=['Masculino', 'Feminino'])

plt.show()
```

Gráfico 6:
O gráfico de quantidade de passageiros por sexo revela que havia mais homens do que mulheres a bordo.

```python
# Quantidade de sobreviventes
plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=df_titanic, palette='cool')
plt.title('Quantidade de Sobreviventes e Não Sobreviventes')
plt.xlabel('Sobreviveu')
plt.ylabel('Quantidade')
plt.xticks(ticks=[0, 1], labels=['Morreu', 'Sobreviveu'])

plt.show()
```

Gráfico 7: O gráfico de quantidade de sobreviventes e não sobreviventes mostra que menos da metade dos passageiros sobreviveu ao desastre.



```python
# Frequência das idades
plt.figure(figsize=(10, 6))
sns.histplot(df_titanic['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Frequência de Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

Gráfico 8: A distribuição de idade dos passageiros é mais concentrada entre 20 e 40 anos, com um pico em torno dos 30 anos.



```python
# Frequência de Fare
plt.figure(figsize=(10, 6))
sns.histplot(df_titanic['Fare'], bins=30, kde=True, color='salmon')
plt.title('Diferença entre o número de passagens compradas baseado no preço')
plt.xlabel('Valor da Passagem')
plt.ylabel('Passagens Compradas')
plt.show()
```

Gráfico 9: A distribuição do valor da passagem é altamente assimétrica, com a maioria dos passageiros pagando menos de 50 dólares, mas alguns pagando valores muito altos.



```python

# Frequência de Embarked
plt.figure(figsize=(8, 5))
sns.countplot(x='Embarked', data=df_titanic, palette='spring')
plt.title('Frequência de Embarque por Porto')
plt.xlabel('Porto de Embarque')
plt.ylabel('Frequência')
plt.show()
```

Gráfico 10: A frequência de embarque por porto indica que a maioria dos passageiros embarcou em Southampton (S), seguido por Cherbourg (C) e Queenstown (Q).
